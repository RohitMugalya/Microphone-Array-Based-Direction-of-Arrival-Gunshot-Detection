```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from qbstyles import mpl_style

mpl_style()
```

```python
gunshot_path = "../data/gunshots/1 (1).wav"
noise_path = "../data/non_gunshots/1-137-A-32.wav"

gunshot, sr = librosa.load(gunshot_path, sr=44100)
noise, _ = librosa.load(noise_path, sr=44100)
```

```python
def plot_spectrogram(y, sr, title):
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.show()
```

```python
# --- Plot Waveforms ---
plt.figure(figsize=(12, 8))

# Gunshot Waveform
plt.subplot(2, 2, 1)
librosa.display.waveshow(gunshot, sr=sr, color='r')
plt.title("Gunshot Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

# Gunshot Spectrogram
plt.subplot(2, 2, 2)
D_gun = librosa.amplitude_to_db(np.abs(librosa.stft(gunshot)), ref=np.max)
librosa.display.specshow(D_gun, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title("Gunshot Spectrogram")

# Noise Waveform
plt.subplot(2, 2, 3)
librosa.display.waveshow(noise, sr=sr, color='b')
plt.title("Non-Gunshot Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

# Noise Spectrogram
plt.subplot(2, 2, 4)
D_noise = librosa.amplitude_to_db(np.abs(librosa.stft(noise)), ref=np.max)
librosa.display.specshow(D_noise, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title("Non-Gunshot Spectrogram")

plt.tight_layout()
```
